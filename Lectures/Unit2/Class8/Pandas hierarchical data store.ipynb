```python
import pandas as pd 
```

```python
df=pd.read_excel('../data/superstore.xls')
```

```python
df=df.set_index(['Customer Name','Order Date']).sort_index(level=0)
#making index two columns, not one 
#order which you list the columns matters 


#Because there are merged columns (aaron bergman); this is called panel data
#so, we treat each of these as their own data set
#customer level data sets
```

```python
#we cannot just shift up and analyze each customer 
#individuals have their own buying patterns 
#.shift(), then, does not work

#LEVEL 0 = NAMES 
#LEVEL 1 = ORDER DATE, AND ALL OF THE COLUMNS ON THE SAME LEVEL 

df['Sales'].shift().head(10)
```

```python
(df['Sales'] - df['Sales'].mean()).mean() #close to 0
```

```python
df['Sales'].sum(level=0) #level=0 means its the first level in the hierarchy, which is by custoemr 
```

```python
df['Sales']-df['Sales'].sum(level=0) # 
```

```python
df['Sales'].mean(level=0) #mean of every customer
```

```python
df['Sales'] - df['Sales'].mean(level=0) #taking every sales # with the mean of each person's sales. 
#each customer, each sale they make, minus the average value of all of their purchases. 
```

```python
#most functions DO NOT ACCEPT level as an argument
df.head(10)['Sales'].pct_change() #will not represent customer level growth 
```

```python
#groupby, however, does take level as an argument, so we can group by cust first 
#and then find the %change within each customer dataset
df.groupby(level=0)['Sales'].pct_change().head(10)
```
