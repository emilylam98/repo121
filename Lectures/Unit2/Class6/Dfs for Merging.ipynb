```python
import pandas as pd
```

```python
data1 = {
    'Zone': ['West', 'South', 'West', 'North', 'East', 'South'],
    'Person': ['Jack', 'Jane', 'Jill', 'Jason', 'Jordan', 'Jenna'],
    'Division': ['Sales', 'Marketing', 'Sales', 'Operations', 'Admin', 'Sales']
}

df1 = pd.DataFrame(data1)
df1
```

```python
data2 = {
    'Region'  : ['West', 'East', 'West', 'North', 'East', 'Midwest'],
    'Person'  : ['Jack', 'Jackson', 'Jill', 'Jason', 'Jordan', 'Adam'],
    'Division': ['Sales', 'Management', 'Sales', 'Operations', 'Admin', 'Admin']
}

df2 = pd.DataFrame(data2)
df2
```

```python
#?df1.merge will show more info 

df1.merge(df2)

#if you do not specify, it'll try to find a match on every column
```

```python
?df1.merge
```

```python
#merging on Zone, looking for matching zone columns on the left (left = df1) and region columns on the right (df2)
#this, however, returns every possible combination 
df1.merge(df2, left_on="Zone",right_on="Region")
```

```python
#on will only merge the rows that match exactly in the row they share. 

df1.merge(df2, on="Person")
```

```python
#merge automatically does an inner join, which only takes the ones that match completely 

df1.merge(df2, how='left') #df1 is the left, so it takes all the values of df1 if it doesn't match 
#it will return a null value if it doesn't exist. 
```

```python
#if we did a right join, we would see null values in Zone 

df1.merge(df2, how='right')
```

```python
df3= df1.copy()
```

```python
#concat takes a list of data frames and concatenates them 
#you could take a list of file names and concatenate them directly. 
pd.concat([df1,df3])
```
